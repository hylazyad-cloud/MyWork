```python
# برنامج إدارة المهام اليومية لسطر الأوامر (تفاعلي بدون واجهات)

# مصفوفة لتخزين المهام
tasks = []

def show_menu():
    print("\n" + "="*30)
    print("   مدير المهام اليومية البسيط   ")
    print("="*30)
    print("1. عرض المهام الحالية")
    print("2. إضافة مهمة جديدة")
    print("3. حذف مهمة محددة")
    print("4. خروج من البرنامج")
    print("="*30)

def view_tasks():
    if not tasks:
        print("\n[!] لا توجد أي مهام في القائمة حالياً.")
    else:
        print("\n--- مهامك الحالية ---")
        for index, task in enumerate(tasks, start=1):
            print(f"{index}. {task}")

def add_task():
    new_task = input("\n[+] اكتب المهمة التي تريد إضافتها: ").strip()
    if new_task:
        tasks.append(new_task)
        print(f"✅ تم إضافة: '{new_task}' بنجاح!")
    else:
        print("❌ خطأ: لا يمكن إضافة مهمة فارغة.")

def delete_task():
    view_tasks()
    if tasks:
        try:
            task_num = int(input("\n[-] أدخل رقم المهمة التي تريد حذفها: "))
            if 1 <= task_num <= len(tasks):
                removed = tasks.pop(task_num - 1)
                print(f"🗑️ تم حذف المهمة: '{removed}'")
            else:
                print("❌ خطأ: الرقم الذي أدخلته غير موجود في القائمة.")
        except ValueError:
            print("❌ خطأ: الرجاء إدخال رقم صحيح فقط.")

# الحلقة الرئيسية للبرنامج (Main Loop)
while True:
    show_menu()
    choice = input("الرجاء اختيار إجراء من القائمة (1-4): ").strip()

    if choice == "1":
        view_tasks()
    elif choice == "2":
        add_task()
    elif choice == "3":
        delete_task()
    elif choice == "4":
        print("\n👋 شكراً لاستخدامك البرنامج. بالتوفيق يا هندسة!")
        break
    else:
        print("❌ خيار غير صحيح! الرجاء إدخال رقم بين 1 و 4.")
```
